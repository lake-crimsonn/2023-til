```python

def genIndex(data, num):
    import random
    dataLen = len(data)
    if type(num)!=int: #isinstance(num, int) == False
        print('integer only :', num)
        return
    elif num > dataLen:
        print('index out of range, max range :', dataLen)
        return
    elif num <= 0:
        print('no negative number :', num)
        return
    else:    
        numList = list(set(random.randrange(0,dataLen) for x in range(num))) # list(set())중복 제거
        numList.sort()
        display(data.iloc[numList,:])

genIndex(titanic_df, 3)
# 중복일 때 복원추출, 비복원추출 고민해야 한다
```
