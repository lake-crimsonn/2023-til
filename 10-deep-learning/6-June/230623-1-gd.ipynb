### 경사하강법
- next w : LR * [(y^-y)*x]평균
- next b : LR * [(y^-y)]평균 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(100) # 플로트 100개 생산
Y = 0.2 * X * 0.5
```

```python
plt.figure(figsize=(4, 3))
plt.scatter(X, Y)
plt.show()
```

```python
def plot_predict(yHat, y):
    plt.scatter(X, y)
    plt.scatter(X, yHat)
    plt.figure(figsize=(12, 6))
    plt.show()
```

```python
def print_plot(w, name):
    plt.figure(figsize=(4,3))
    plt.plot(w)
    plt.title(name)
    plt.show()
```

```python
w = np.random.uniform(-1, 1) # -1 ~ 1 사이의 실수를 리턴
b = np.random.uniform(-1, 1)
learning_rate = 1
check = 5
wArr = []
wgradArr = []
for epoch in range(100):
    y_hat = w * X + b # 랜덤의 w, b값으로 시작

    # gradient descending 
    # 평균제곱오차를 미분
    w_grad = ((y_hat - Y)* X).mean() * learning_rate
    b_grad = ((y_hat - Y)).mean() * learning_rate

    # update w, b
    w = w - w_grad
    b = b - b_grad

    wgradArr.append(w_grad)
    wArr.append(w)
    
    # 표 그려주기
    if epoch % check == 0:
        plot_predict(y_hat,Y)

    # 너무 많이 반복하지 않기
    err = np.abs(y_hat - Y).mean()
    if err < 0.001:
        break

print_plot(wArr, 'w')
print_plot(wgradArr, 'w_grad')
```
