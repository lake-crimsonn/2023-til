```python
from glob import glob
from matplotlib import pyplot as plt
import numpy as np
import os
import pandas as pd
import cv2
```

```python
BASE_URL = 'C:\data\celebaMask\dataset'
img_url = os.path.join(BASE_URL,'img')
name_list = []
for i in os.listdir(img_url):
    name_list.append(int(i.split('.')[0]))
name_list.sort()
img_url_lst = []
for i in name_list:
    img_url_lst.append(os.path.join(img_url,str(i)+'.jpg'))

anno_url = os.path.join(BASE_URL,'mask')
anno_url_list = []
for i in os.listdir(anno_url):
    anno_url_list.append(os.path.join(anno_url,i))
anno_url_list
```

```python
img = cv2.imread(img_url_lst[0])
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
anno_img = cv2.imread(anno_url_list[0])
anno_img = cv2.cvtColor(anno_img, cv2.COLOR_BGR2RGB)
plt.imshow(anno_img)
```
