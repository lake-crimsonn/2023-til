```python
import cv2 
from glob import glob
import numpy as np
"""   """
dataset_dir = 'C:\\data\\celebaMask\\dataset'
png_list = glob(dataset_dir+'\\*.png') # glob , *.png png   , os.listdir 

img1 = cv2.imread(png_list[0], cv2.IMREAD_GRAYSCALE)
img1 = cv2.resize(img1, (512, 512))
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2BGRA)

img2 = cv2.imread(png_list[1], cv2.IMREAD_GRAYSCALE)
img2 = cv2.resize(img2, (512, 512))
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2BGRA)

img3 = cv2.imread(png_list[2], cv2.IMREAD_GRAYSCALE)
img3 = cv2.resize(img3, (512, 512))
img3 = cv2.cvtColor(img3,cv2.COLOR_BGR2BGRA)

print(np.shape(img1), np.shape(img2), np.shape(img3))

for i in range(len(img1)):
    for j in range(len(img1[0])):
        if (img1[j][i] == [0,0,0,255]).all():
            img1[j][i] = [0,0,0,0]
            
for i in range(len(img2)):
    for j in range(len(img2[0])):
        if (img2[j][i] == [0,0,0,255]).all():
            img2[j][i] = [0,0,0,0]
            
for i in range(len(img3)):
    for j in range(len(img3[0])):
        if (img3[j][i] == [0,0,0,255]).all():
            img3[j][i] = [0,0,0,0]
            
mimg = img1 + img2 + img3
mimg = cv2.cvtColor(mimg, cv2.COLOR_BGRA2BGR)
print(np.shape(mimg))
print(mimg)
# cv2.imshow('img',mimg)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
cv2.imwrite('C:\data\celebaMask\dataset\img.png',mimg)
```

```python
import cv2 
from glob import glob
import numpy as np

"""         """
dataset_dir = 'C:\\data\\celebaMask\\dataset\\mask'

ears_list = glob(dataset_dir+'\\*ear.png') # glob , *.png png   , os.listdir 
skin_list = glob(dataset_dir+'\\*skin.png')
ears_list, skin_list

skin_dict = {}
for i in skin_list:
    skin_id = i.split('\\')[-1].split('_')[0]
    skin_address = i
    skin_dict[skin_id] = skin_address
    
file_dict = {}
for i in ears_list:
    ears_id = i.split('\\')[-1].split('_')[0]
    ears_address = i
    if ears_id not in file_dict:
        file_dict[ears_id] = []
    file_dict[ears_id].append(ears_address)
        
for i in skin_dict.items():
    if i[0] in file_dict.keys():
        file_dict[i[0]].append(i[1])
file_dict
```

```python
len(file_dict)
```

```python
img_dict = {}

for filename in list(file_dict.keys())[:5000]: 
    for img in file_dict[filename]:
        img = cv2.imread(img, cv2.IMREAD_UNCHANGED)
        img = cv2.resize(img, (512, 512))
        if filename not in img_dict:
            img_dict[filename] = []
        img_dict[filename].append(img)
```

```python
for filename in list(file_dict.keys())[:5000]:
    
    imgs = img_dict[filename]
    
    for k in imgs:
        for i in range(len(k)):
            for j in range(len(k[0])):
                if (k[j][i] == [255,255,255]).all():
                    k[j][i] = [1,1,1]
    
    mimg = imgs[0]
    for img in imgs:
        mimg = cv2.bitwise_or(mimg,img) # 1 or 1 = 1, 1 or 2 = 3
    
    cv2.imwrite(f'C:\data\celebaMask\dataset\skin\{filename}_skin.png',mimg)
    len(os.listdir('C:\data\celebaMask\dataset\skin'))
```
