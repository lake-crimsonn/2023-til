```python
import cv2 
from glob import glob
import numpy as np
import pandas as pd

dataset_dir = "C:\data\celebaMask\dataset\mask"
img_wh = 512

mask_addr_list = glob(dataset_dir+'\\*neck.png') # glob 장점, *.png png파일만 가져올 수 있음, os.listdir 오류남

mask_list = {}
mask_name_list = []
for i in mask_addr_list:
    name = i.split('\\')[-1].split('_')[0]
    mask_name_list.append(name)
    mask_list[name] = i

mask_df = pd.DataFrame(zip(mask_list,mask_addr_list), columns=['name','addr'])

from sklearn.model_selection import train_test_split

# test_size 명시하지 않으면 0.25
train, test = train_test_split(mask_df,test_size=0.1, random_state=777)
train, validation = train_test_split(train, test_size=0.1, random_state=777)

# 인덱스 초기화
train.reset_index(inplace=True, drop=True)
validation.reset_index(inplace=True, drop=True)
test.reset_index(inplace=True, drop=True)

len(train), len(validation), len(test)

```

```python
# 저장
train.to_csv('C:\data\celebaMask\dataset\\neck3_train.csv', header=False)
test.to_csv('C:\data\celebaMask\dataset\\neck3_test.csv', header=False)
validation.to_csv('C:\data\celebaMask\dataset\\neck3_valid.csv', header=False)
```

```python
import os
img_addr_list = sorted(glob('C:\data\celebaMask\dataset\img\\*.jpg'))
find_df = validation

img_name_dict = {}
for i in img_addr_list:
    img_name = i.split('\\')[-1].split('.')[0]
    img_name_dict[img_name] = i
    
img_move_list = []
for i in range(len(find_df)):
    img_name = find_df.iloc[i,0]
    
    if img_name[0] == '0':
        img_name = str(img_name).lstrip('0')      
        
    if img_name in img_name_dict.keys():
        img_move_list.append(img_name_dict[img_name])


""" 복사하기 """
import shutil
copyfiles = img_move_list
dest = 'C:\data\celebaMask\dataset\\neck2\\val\images'
for i in copyfiles:
    shutil.copy(i,dest)
```
