```python
neck_addr_list = glob('C:\data\celebaMask\dataset\\neck_255\\*.png')
len(neck_addr_list) # 851   
img_wh = 512

for i in neck_addr_list:
    filename = i.split('\\')[-1]
    img = cv2.imread(i,cv2.IMREAD_UNCHANGED)
    img = cv2.resize(img, (img_wh,img_wh))
    for j in range(len(img)):
        for k in range(len(img[0])):
            if (img[k][j] == [255,255,255]).all():
                img[k][j] = [1,1,1]

    cv2.imwrite(f"C:\data\celebaMask\dataset\\neck_1\\{filename}", img)
```
