```python
original_dataset_dir = 'C:\data\celebaMask\CelebAMask-HQ'
img_folder_list = os.listdir(original_dataset_dir+'/CelebA-HQ-img')
folder_list = os.listdir(original_dataset_dir)
anno_folder_list = os.listdir(os.path.join(original_dataset_dir,'CelebAMask-HQ-mask-anno'))

""" 이름이 0~14폴더 정렬시키기"""
num_arr = anno_folder_list.copy()
for idx, e in enumerate(anno_folder_list):
    num_arr[idx] = int(e)
num_arr.sort()
for idx, e in enumerate(num_arr):
    num_arr[idx] = str(e)
num_arr

""" 각 숫자폴더에 있는 파일리스트 받아오기 """
anno_folder_list = os.path.join(original_dataset_dir,'CelebAMask-HQ-mask-anno')
anno_file_list = []
for e in num_arr:
    anno_file_list.append(os.listdir(os.path.join(anno_folder_list, e)))

anno_file_list[:400]

""" 0~14 폴더에서 마스크 이름 가져오기 """
find_list = ['skin.png','neck.png','hair.png','ear.png'] # 특정 마스크
anno_list = [] # 숫자 폴더 주소 리스트
for e in num_arr:
    anno_list.append(os.path.join(anno_folder_list, e))
found_list = []
for e in anno_list:
    ls = os.listdir(e) # 숫자 폴더의 이미지 리스트
    fl = []
    for i in ls:
        if i.split('_')[-1] in find_list:
            fl.append(i)
    found_list.append(fl)
found_list
```

```python
""" 복사 절대주소 만들기 """
copy_files = []
for idx, al in enumerate(anno_list):
    for fl in found_list[idx]:
        copy_files.append(os.path.join(al, fl))
        
copy_files
```

```python
import cv2 
from glob import glob
import numpy as np
""" 얼굴 귀 합치기"""
dataset_dir = 'C:\\data\\celebaMask\\dataset'
png_list = glob(dataset_dir+'\\*.png') # glob 장점, *.png png파일만 가져올 수 있음, os.listdir 오류남

img1 = cv2.imread(png_list[0], cv2.IMREAD_GRAYSCALE)
img1 = cv2.resize(img1, (512, 512))
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2BGRA)

img2 = cv2.imread(png_list[1], cv2.IMREAD_GRAYSCALE)
img2 = cv2.resize(img2, (512, 512))
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2BGRA)

img3 = cv2.imread(png_list[2], cv2.IMREAD_GRAYSCALE)
img3 = cv2.resize(img3, (512, 512))
img3 = cv2.cvtColor(img3,cv2.COLOR_BGR2BGRA)

print(np.shape(img1), np.shape(img2), np.shape(img3))

for i in range(len(img1)):
    for j in range(len(img1[0])):
        if (img1[j][i] == [0,0,0,255]).all():
            img1[j][i] = [0,0,0,0]
            
for i in range(len(img2)):
    for j in range(len(img2[0])):
        if (img2[j][i] == [0,0,0,255]).all():
            img2[j][i] = [0,0,0,0]
            
for i in range(len(img3)):
    for j in range(len(img3[0])):
        if (img3[j][i] == [0,0,0,255]).all():
            img3[j][i] = [0,0,0,0]
            
mimg = img1 + img2 + img3
mimg = cv2.cvtColor(mimg, cv2.COLOR_BGRA2BGR)
print(np.shape(mimg))
print(mimg)
# cv2.imshow('img',mimg)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
cv2.imwrite('C:\data\celebaMask\dataset\img.png',mimg)
```

```python
import cv2 
from glob import glob
import numpy as np

""" 얼굴과 귀가 있는 파일 불러와서 하나의 딕셔너리로 정리하기 """
dataset_dir = 'C:\\data\\celebaMask\\dataset\\mask'

ears_list = glob(dataset_dir+'\\*ear.png') # glob 장점, *.png png파일만 가져올 수 있음, os.listdir 오류남
skin_list = glob(dataset_dir+'\\*skin.png')
ears_list, skin_list

skin_dict = {}
for i in skin_list:
    skin_id = i.split('\\')[-1].split('_')[0]
    skin_address = i
    skin_dict[skin_id] = skin_address
    
file_dict = {}
for i in ears_list:
    ears_id = i.split('\\')[-1].split('_')[0]
    ears_address = i
    if ears_id not in file_dict:
        file_dict[ears_id] = []
    file_dict[ears_id].append(ears_address)
        
for i in skin_dict.items():
    if i[0] in file_dict.keys():
        file_dict[i[0]].append(i[1])
file_dict
```

```python
hair = "C:\data\celebaMask\dataset\\00002_hair.png"
skins = "C:\\data\\celebaMask\\dataset\\skin\\00002_skin.png"
neck = "C:\\data\\celebaMask\\dataset\\00002_neck.png"

img1 = cv2.imread(hair, cv2.IMREAD_UNCHANGED)
img1 = cv2.resize(img1, (512, 512))

img2 = cv2.imread(skins, cv2.IMREAD_UNCHANGED)
img2 = cv2.resize(img2, (512, 512))

img3 = cv2.imread(neck, cv2.IMREAD_UNCHANGED)
img3 = cv2.resize(img3, (512, 512))

# 헤어 컬러 변경
for i in range(len(img1)):
    for j in range(len(img1[0])):
        if (img1[j][i] == [255,255,255]).all():
            img1[j][i] = [2,2,2]
            
mimg = cv2.bitwise_or(img1,img2)

# 목과 헤어 겹치는 부위 수정 
for i in range(len(mimg)):
    for j in range(len(mimg[0])):
        if (mimg[j][i] == [3,3,3]).all():
            mimg[j][i] = [2,2,2]

# 목 컬러 변경
for i in range(len(img3)):
    for j in range(len(img3[0])):
        if (img3[j][i] == [255,255,255]).all():
            img3[j][i] = [3,3,3]

# 합치기
mimg = cv2.bitwise_or(mimg,img3)

# cv2.imshow("img",mimg)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

cv2.imwrite(f'C:\data\celebaMask\dataset\{filename}_m.png',mimg)
```
