# 파일 리스트를 csv로 변환하기

```python
import numpy as np
import pandas as pd
from glob import glob
```

```python
total = glob('dataset/*/*/*')
train = glob('dataset/train/*/*')
test = glob('dataset/test/*/*')
val = glob('dataset/val/*/*')

len(total), len(train), len(test), len(val)
```

```python
splitted_arr = []
for t in train:
    splited = t.split('\\')
    splited = splited[1:]
    splitted_arr.append(splited)

train_df = pd.DataFrame(data=splitted_arr, columns=['case','filename'])

train_df
```

```python
splitted_arr = []
for t in test:
    splited = t.split('\\')
    splited = splited[1:]
    splitted_arr.append(splited)

test_df = pd.DataFrame(data=splitted_arr, columns=['case','filename'])

test_df
```

```python
splitted_arr = []
for t in val:
    splited = t.split('\\')
    splited = splited[1:]
    splitted_arr.append(splited)

val_df = pd.DataFrame(data=splitted_arr, columns=['case','filename'])

val_df
```

```python
len(train_df), len(test_df), len(val_df)
```

```python
train_df.to_csv('train.csv', index=True)
test_df.to_csv('test.csv', index=True)
val_df.to_csv('val.csv', index=True)
```
