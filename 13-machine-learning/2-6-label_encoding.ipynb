## 레이블인코딩
image

- 주로 y값 인코딩에 사용한다
- 원핫인코딩을 하기 전에 숫자값 변환을 위해 사용하기도 한다

```python
from sklearn.preprocessing import LabelEncoder

items = ('TV, 냉장고, 전자렌지, 컴퓨터, 선풍기, 선풍기, 믹서, 믹서').split(', ')
print('items :',items)

encoder = LabelEncoder()
encoder.fit(items) # 중복을 제거하고 0,1,2,3,4와 같은 번호를 갖는다
print('encoder_classes :',encoder.classes_) # 중복을 제거하고 오름차순 정렬
labels = encoder.transform(items)
print('인코딩 변환값 :', labels) # 0-TV, 1-냉장고, 4-'전자렌지', ...
print('디코딩 원본 값 :',encoder.inverse_transform([5,5,4,4,3,3,2,2,1,1]))
```
