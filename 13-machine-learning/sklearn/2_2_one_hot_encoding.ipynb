## 원핫인코딩
image
- 카테고리 변수가 3개 이상일 때 사용한다
- 하나의 어레이가 들어간다, 레이블인코딩은 하나하나 필드단위로 들어간다
- 1차원 배열은 사용하지 못하기 때문에 2차원 배열로 만들어줘야 한다. reshape(-1,1)
- 교재 118p

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
items = ('TV, 냉장고, 전자렌지, 컴퓨터, 선풍기, 선풍기, 믹서, 믹서').split(', ')
print(items)

# 숫자 변환
encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items) # encoder.inverse_transform()로 디코딩 가능
print('숫자로 변환 :',labels) 

# 1차원 -> 2차원 배열
items = np.array(items).reshape(-1,1)

# 원핫인코딩 핏
oh_encoder = OneHotEncoder()
oh_encoder.fit(items) # 중복제거하고 오름차순 정렬
print('핏 카테고리 보기 :',oh_encoder.categories_) # 카테고리보기

# 원핫인코딩 트랜스폼
oh_label = oh_encoder.transform(items)
print('원핫인코딩 적용한 변수의 타입 :',type(oh_label))
print(oh_label.toarray()) #칼럼: 카테고리, 로우: items
print('희소행열 크기 :',oh_label.shape) # sparse matrix : 희소행렬 (0이 많은 행렬)
```
