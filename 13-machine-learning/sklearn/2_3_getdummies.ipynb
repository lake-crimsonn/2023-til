#### get_dummies()로 원핫인코딩이 가능함
- 카테고리에 해당하는 문자변수를 낱개의 필드(x)로 쪼갠다
- 유니크한 카테고리 낱개들의 갯수가 너무 많으면 x변수로 다 처리가 되지 않는다
- 라쏘 모듈을 사용하면 된다(L1 정규화)
- 교재 119p

```python
import pandas as pd

df = pd.DataFrame({'items':('TV, 냉장고, 전자렌지, 컴퓨터, 선풍기, 선풍기, 믹서, 믹서').split(', ')})
df.head()
```

```python
pd.get_dummies(df)
```
