```python
################################################33
#### 얼굴인식 하는 xml자료 연동하기
#### 이미지 자료 읽기
###########################################
import cv2

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_alt.xml') #정면인식하는 코드
```

```python
def detect(frame, sysCnt):
    cnt = 0
    faces = face_cascade.detectMultiScale(frame, scaleFactor=1.1, minNeighbors=5, minSize=(5,5))
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x-20, y-20), (x + w+20, y + h+20), (255, 0, 0), 2)
        rect_arr = frame[y-20:y + h+20, x-20:x + w+20]
        cnt+=1
        cv2.putText(frame, str(cnt) , (x-10,y-30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 2)
        if cnt >= 2 and sysCnt%30:
            cv2.imwrite('./saveImg/'+ str(cnt) +'.png', rect_arr)
        
    cv2.imshow("new",frame)
```

```python
import cv2
webcam = cv2.VideoCapture(0)
webcam.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
webcam.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
cnt = 0
sysCnt = 0

if not webcam.isOpened():
    print("Could not open webcam")
    exit()

while webcam.isOpened():
    status, frame = webcam.read()
    sysCnt += 1
    if status:
        cv2.imshow("new", frame)
        detect(frame, sysCnt)   
        print(sysCnt)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    if cv2.waitKey(1) == ord('a'):
      fileName='frame' + str(cnt) + '.jpg'
      cv2.imwrite('c:/data/cascade_cv2/'+fileName, frame)
      print('a키가 눌려서 저장되었습니다.')
      cnt+=1
      

webcam.release()
cv2.destroyAllWindows()
```
