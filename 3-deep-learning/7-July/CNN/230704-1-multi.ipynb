```python
from glob import glob
baseFolder = 'C:/data/kfood/*'
Folders = glob(baseFolder)
dataList = []

for folder in Folders:
    subFolders = glob(folder+'/*')
    for subfolder in subFolders:
        fileList = glob(subfolder+'/*.jpg')
        dataList.append(fileList)
        
dataList
```

```python
import numpy as np
np.shape(dataList)
```

```python
import pandas as pd
df = pd.DataFrame()
for list in dataList:
    df = pd.concat([df,pd.DataFrame(list)])
df
```

```python
df.isna().sum()
```

```python
df.iloc[:,0].str.split('\\')[1] # index 1
df.iloc[:,0].str.split('\\').str[1] # columns 1
df.iloc[:,0].str.split('\\') # Series to string
```

```python
df['super'] = df.iloc[:,0].str.split('\\').str[1]
df['sub'] = df.iloc[:,0].str.split('\\').str[2]
df['name'] = df.iloc[:,0].str.split('\\').str[3]
df
```

```python
tmp = pd.get_dummies(df[['super','sub']], dtype='uint8', prefix='',prefix_sep='')
tmp
```

```python
df = df.drop(columns=['super','sub','name'])
```

```python
df = df.rename(columns={0:'image'})
```

```python
df3 = pd.concat([df,tmp], axis=1)
df3.reset_index(inplace=True, drop=True)
```

```python
from sklearn.model_selection import train_test_split

#     0.25
train, test = train_test_split(df3, random_state=777)
validation, train = train_test_split(train, random_state=777)

validation
```

```python
train.reset_index(inplace=True, drop=True)
validation.reset_index(inplace=True, drop=True)
test.reset_index(inplace=True, drop=True)
```

```python
train.to_csv('c:/data/kfood2/train.csv', header=False)
test.to_csv('c:/data/kfood2/test.csv', header=False)
validation.to_csv('c:/data/kfood2/valid.csv', header=False)
```
